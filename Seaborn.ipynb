# size

```python
sns.lineplot(x="bill_length_mm",y='bill_depth_mm',data=cs_df,hue="sex",size=50)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y='bill_depth_mm',data=cs_df,hue="sex",size=50,palette="Accent")
plt.show()
```
